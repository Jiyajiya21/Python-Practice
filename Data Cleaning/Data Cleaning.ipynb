```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r"/Users/jiyadariyani/Documents/Projects/Python Projects/Customer_Call_List.csv")
df
```

```python
df = df.drop(columns = "Unnamed: 9")
df
```

```python
# Removing duplicates
df = df.drop_duplicates()
df
```

```python
#Standardizing Data
df = df.copy()
df['Last_Name'] = df['Last_Name'].str.strip("./_")
df
```

```python
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
#str(df["Phone_Number"])
#Standardizing the Columns
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
#df["Phone_Number"] = df["Phone_Number"].str.replace('nan','', regex=False)
#df["Phone_Number"] = df["Phone_Number"].str.replace('Na','', regex=False)
str(df["Phone_Number"])
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#Removing Unwanted rows
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','', regex=False)
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','', regex=False)
df
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('Yes', 'Y')
```

```python
df["Do_Not_Contact"]
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('No', 'N')
```

```python
df["Paying Customer"] = df["Paying Customer"].replace('Yes', 'Y')
```
