## Grouping the Numerical Data 

```python
df_numeric = df[['2022 Population', '2020 Population', '2000 Population', 
                 '2010 Population', '2015 Population', '1970 Population', 
                 '1980 Population', '1990 Population', 'World Population Percentage']]

# Calculate the mean for each group in the 'Continent' column
df_grouped2 = df_numeric.groupby(df['Continent']).mean().sort_values(by='World Population Percentage', ascending=False)
df_grouped2
```

## Visualizing the Data By using the Heat Maps

```python
# Set up the matplotlib figure with the desired size
plt.figure(figsize=(10, 7)) # Fig Size is the size of the box for plotting the Visual
df_grouped = df.corr(numeric_only=True) # taking only the Numerical Data 
sns.heatmap(df_grouped, annot=True, cmap='coolwarm') # annot specifies the Numerics at each cell
```
